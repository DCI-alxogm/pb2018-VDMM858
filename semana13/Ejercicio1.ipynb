```python
import matplotlib.pyplot as plt
import numpy as np 
%matplotlib inline 
```

```python
# Esto hace que se generen 9 puntos aleatorios 
x=np.random.rand(9)
y=np.random.rand(9)
z=np.array([x,y])
plt.scatter(x,y,color='green')
```

```python
# Con este for se imprimen 100 puntos en forma de círculo, los cuales van a rodear aleatoriamente a los puntos
for i in range(0,100,1):
        y=(0.1*np.sin(i))
        x=(0.1*np.cos(i))
        plt.scatter(x,y,color='purple')
```

```python
# Con esto se rodean 3 puntos aleatorios 
for j in range (10):
    x=np.random.rand(3)
    y=np.random.rand(3)
    z=np.array([x,y])
    a=z[0,j]
    b=z[1,j]
    plt.scatter(x,y,color='green')
    for k in range(0,100,1):
            y=((0.1)*np.sin(k))
            x=((0.1)*np.cos(k))
            plt.scatter(x+a,y+b,color='purple')
```
