```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
## Definir el valor de las variables 
w=1 ## Velocidad angular 
h=0.01 ## Variación en el tiempo 
tmax=60 ## Tiempo máximo (en segundos)
vi=0.5 ## Velocidad inicial 
```

```python
Ti=np.arange(0,tmax,h)
v=np.zeros(1000)
v[0]=vi
theta=np.zeros(1000)
theta[0]=10
i=np.arange(0,1000,h)
```

```python
for i in range(0,999,1):
    theta[i+1]=theta[i]+h*v[i]
    v[i+1]=v[i]+h*(-w**2*theta[i])
```

```python
plt.plot(v,'green')
plt.xlabel("Velocidad")
```

```python
plt.plot(theta,'blue')
plt.xlabel('theta')
```

```python
plt.plot(v,theta,'red')
plt.xlabel("velocidad")
plt.ylabel("theta")
```
